### Extract valid information from the dataframe 

```python
df_motor = df_blank.copy()
df_motor = df_motor.rename(columns = {'Minimum Rating' : 'Minimum_Rating_in_eKW', 'Maximum Rating' : 'Maximum_Rating_in_eKW', 'Voltage': 'Voltage_in_Volts', 'Frequency': 'Frequency_in_Hz', 'Speed': 'Speed_in_rpm'})

def extract_min_max_rating(temp):
    value = re.search(r"\d+", temp)
    return(value.group(0))

def extract_volt(temp):
    if re.search(r".*(Volts$|volts$)", str(temp)):
        if re.search(r"([0-9]+)\s?(?=(to|-))", temp):
            value = re.search(r"([0-9]+)\s?(?=(to|-))", temp)
            return(value.group(1))
        else:
            value = re.search(r"(.*)(Volts$|volts$)", str(temp))
            return(value.group(1))

    elif re.search(r".*([0-9]+$)", str(temp)):
        if re.search(r"([0-9]+)\s(?=to)", temp):
            value = re.search(r"([0-9]+)\s(?=to)", temp)
            return(value.group(1))
        else:
            value = re.search(r"(.*)(\d+$)", str(temp))
            return(value.group(1))
    else:
        return(np.NaN)
    
def extract_freq(temp):
    if re.search(r".*(Hz$)", temp):
        if re.search(r"([0-9]+)\s(?=or)", temp):
            value = re.search(r"([0-9]+)\s(?=or)", temp)
            return(value.group(1))
        else:
            value = re.search(r"(.*)(Hz$)", temp)
            return(value.group(1))
        
    elif re.search(r".*([0-9]+$)", temp):
        value = re.search(r"(\d+)\s(?=Hz)", temp)
        return(value.group(1))  
    else:
        return(np.NaN)

def extract_speed(temp):
    if re.search(r".*(rpm$|RPM$)", temp):
        if re.search(r"([0-9]+)\s(?=or)", temp):
            value = re.search(r"([0-9]+)\s(?=or)", temp)
            return(value.group(1))
        else:
            value = re.search(r"(.*)(rpm$|RPM$)", temp)
            return(value.group(1))
    elif re.search(r"(\d+)\s(?=(rpm$|RPM$))", temp):
        value = re.search(r"(\d+)\s(?=(rpm$|RPM$))", temp)
        return(value.group(1))  
    else:
        return(np.NaN)     
        
```

```python
df_motor.Minimum_Rating_in_eKW = df_motor.Minimum_Rating_in_eKW.apply(extract_min_max_rating)
df_motor.Maximum_Rating_in_eKW = df_motor.Maximum_Rating_in_eKW.apply(extract_min_max_rating)

df_motor.Frequency_in_Hz = df_motor.Frequency_in_Hz.apply(extract_freq)
df_motor.Voltage_in_Volts = df_motor.Voltage_in_Volts.apply(extract_volt)
df_motor.Speed_in_rpm = df_motor.Speed_in_rpm.apply(extract_speed)
```
